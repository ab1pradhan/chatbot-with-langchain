```python

# Initialize message lists
innovator_system_prompt = {
    "role": "system",
    "content": "You are an intelligent and visionary entrepreneur. You propose innovative business ideas."
}

critic_system_prompt = {
    "role": "system",
    "content": "You are a sharp, analytical critic. You identify flaws, challenges, and areas of improvement in business ideas."
}

# Initial trigger
initial_prompt = {
    "role": "user",
    "content": "Propose a new startup idea that solves a common productivity problem for remote teams."
}

# Chat history for both models
conversation = [innovator_system_prompt, initial_prompt]

# Run 5 back-and-forth turns
for turn in range(5):
    print(f"\n🔁 Round {turn + 1}\n")

    # ---- Innovator generates idea or responds with improvement ----
    response_idea = openai_client.chat.completions.create(
        model="gpt-4o-mini",
        messages=conversation
    )
    idea = response_idea.choices[0].message.content
    print("🧠 Innovator:", idea)

    # Append Innovator response to conversation
    conversation.append({"role": "assistant", "content": idea})

    # ---- Critic gives feedback on the idea ----
    critic_prompt = {"role": "user", "content": f"Analyze and find issues in the following idea:\n\n{idea}"}
    critic_convo = [critic_system_prompt, critic_prompt]

    response_critic = openai_client.chat.completions.create(
        model="gpt-4o-mini",
        messages=critic_convo
    )
    critique = response_critic.choices[0].message.content
    print("🧐 Critic:", critique)

    # Feed critique back to innovator
    conversation.append({"role": "user", "content": f"Here are the problems raised:\n{critique}\nImprove the idea accordingly."})

print("\n✅ Conversation Complete.")
```
